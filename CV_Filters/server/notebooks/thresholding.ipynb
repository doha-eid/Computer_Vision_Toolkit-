```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
```

```python
# img= mpimg.imread('shapes-01.jpg')
img= mpimg.imread('..//static//assests//image1.jpg')
plt.imshow(img)
plt.show()
```

```python
#function gray scale 
def rgbtogray(image):
    r,g,b=image[:,:,0],image[:,:,1],image[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray
```

```python
grayScale=rgbtogray(img)
plt.imshow(grayScale,cmap="gray")
plt.show()
```

```python
def globalThresholding(img, n):
    img_shape = img.shape
    height = img_shape[0]
    width = img_shape[1]
    for row in range(width):
        for column in range(height):
            if img[column, row] > n:
                img[column, row] = 0
            else:
                img[column, row] = 255
    return img
```

```python
image=globalThresholding(grayScale,200)
plt.imshow(image,cmap="gray")
plt.show()
```

```python
# global thresholding
min_rgb_value = 0
max_rgb_value = 255
globalImage = np.zeros_like(img)
globalImage[grayScale >= max_rgb_value // 2] = max_rgb_value
plt.imshow(globalImage, cmap="gray")
plt.show()
```

```python
def localThresholding(img, radius):
    image = np.zeros_like(img)
    max_row, max_col = img.shape
    for i, row in enumerate(img):
        y_min = max(0, i - radius)
        y_max = min(max_row, i + radius + 1)
        for j, elem in enumerate(row):
            x_min = max(0, j - radius)
            x_max = min(max_col, j + radius + 1)
            window = img[y_min:y_max, x_min:x_max]
            if img[i, j] >= np.median(window):
                image[i, j] = 255
    return image
```

```python
local= localThresholding(grayScale,40)
plt.imshow(local, cmap="gray")
plt.show()
```
